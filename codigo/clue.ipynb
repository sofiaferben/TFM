```python
import pandas as pd 
file_path="C:/Users/Usuario/Desktop/TFM/datos/CLUE_data/repurposing_samples_20180907.txt"

# Read the file and remove metadata lines (those starting with "!")
with open(file_path, "r", encoding="utf-8") as file:
    lines = file.readlines()

# Filter out lines that start with "!"
data_lines = [line.strip() for line in lines if not line.startswith("!") and line.strip()]

# Convert to DataFrame
df_reporpusing_samples = pd.DataFrame([line.split("\t") for line in data_lines[1:]], columns=data_lines[0].split("\t"))
```

```python
import pandas as pd 
file_path="C:/Users/Usuario/Desktop/TFM/datos/CLUE_data/repurposing_drugs_20180907.txt"

# Read the file and remove metadata lines (those starting with "!")
with open(file_path, "r", encoding="ISO-8859-1") as file:
    lines = file.readlines()

# Filter out lines that start with "!"
data_lines = [line.strip() for line in lines if not line.startswith("!") and line.strip()]

# Convert to DataFrame
df_reporpusing_drugs = pd.DataFrame([line.split("\t") for line in data_lines[1:]], columns=data_lines[0].split("\t"))
```

Contains annotation-level information including compound name, clinical phase, mechanism of action, and protein target. No entiedno mucho las variables.
* Moa : mechanism of action
* target: target gene
* disease_area: la area del disease que puede ser interesante

```python
df_reporpusing_drugs
```

```python
# Get the unique values in the "disease_area" column
unique_disease_areas = df_reporpusing_drugs["disease_area"].dropna().unique()
```

```python
print(type(unique_disease_areas))
```

```python
unique_disease_areas
```

```python
# Filter the DataFrame to include only rows where "disease_area" contains "neurology/psychiatry"
df_neuro_psychiatry = df_reporpusing_drugs[df_reporpusing_drugs["disease_area"].str.contains("neurology/psychiatry", na=False, case=False)]
```

```python
df_neuro_psychiatry
```

```python
# Filter df_repurposing_samples to keep only rows where pert_iname matches df_neuro_psychiatry
df_matched_samples = df_reporpusing_samples[
    df_reporpusing_samples["pert_iname"].isin(df_neuro_psychiatry["pert_iname"])
]
```

```python
df_matched_samples
```

hay por cada drug diferente vendor y purity  tal , para matcehar nos quedamos con el primero no lo mejor 

```python
df_matched_samples_unique = df_matched_samples.drop_duplicates(subset=["pert_iname"])
```

```python
# Merge df_neuro_psychiatry with df_matched_samples on "pert_iname"
df_neuro_psychiatry_enriched = df_neuro_psychiatry.merge(
    df_matched_samples_unique[["pert_iname", "broad_id", "expected_mass", "smiles", "InChIKey", "pubchem_cid"]],
    on="pert_iname",
    how="left"  # Keeps all rows from df_neuro_psychiatry, adding NaN if no match
)
```
