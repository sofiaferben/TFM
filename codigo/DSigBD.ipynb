```python
import pandas as pd 
file_path="C:/Users/Usuario/Desktop/TFM/datos/DSigBD/DSigDB_All.gmt.txt"

with open(file_path, "r") as file:
    lines = file.readlines()

# Process lines to split on tabs
data = [line.strip().split("\t") for line in lines]

# Convert into a DataFrame
df_all_gmt = pd.DataFrame(data)
```

```python
df_all_gmt
```

juntamos y lo guardamos en un csv que usar en el futuro oara hacer pues algo 

```python
# First, ensure the column names for df_all_gmt
df_all_gmt.columns = ["Drug", "URL"] + [f"Target_{i}" for i in range(1, df_all_gmt.shape[1] - 1)]
```

```python
df_all_gmt
```

```python
# Merge df_all_info_detailed with df_all_gmt on the Drug_Name column
df_merged = df_all_info_detailed.merge(df_all_gmt[["Drug", "URL"]], on="Drug", how="left")
```

```python
df_merged
```

```python
df_merged.to_csv("C:/Users/Usuario/Desktop/TFM/datos/DSigBD/sig_detailed_url.csv", index=False)
```
